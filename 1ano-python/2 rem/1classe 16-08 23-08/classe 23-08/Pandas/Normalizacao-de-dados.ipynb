##### Formula da normalização de dados: x - min / max - min (max,min)

```python
import pandas as pd
```

```python
v1 = [200, 100, 1000, 150]
v2 = [10, 20, 30, 40]
```

```python
v1 = pd.Series([200, 100, 1000, 150], name = 'Series 1')

v1_min = v1.min()
v1_max = v1.max()

(v1 - v1_min) / (v1_max - v1_min)


```

```python
v2 = pd.Series([10, 20, 30, 40], name = 'Series 2')

v2_min = v2.min()
v2_max = v2.max()

(v2 - v2_min) / (v2_max - v2_min)
```

```python
import matplotlib.pyplot as plt

# Defining the Series
v2 = pd.Series([10, 20, 30, 40], name='Series 2')

# Normalizing the Series
v2_min = v2.min()
v2_max = v2.max()
normalized_v2 = (v2 - v2_min) / (v2_max - v2_min)

# Plotting the normalized values
plt.figure(figsize=(8, 6))
plt.plot(v2.index, normalized_v2, marker='o', linestyle='-', color='b', label='Normalized Series')
plt.title('Normalized Series 2')
plt.xlabel('Index')
plt.ylabel('Normalized Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Defining the Series
v1 = pd.Series([200, 100, 1000, 150], name = 'Series 1')

# Normalizing the Series
v1_min = v1.min()
v1_max = v1.max()
normalized_v1 = (v1 - v1_min) / (v1_max - v1_min)

# Plotting the normalized values
plt.figure(figsize=(8, 6))
plt.plot(v1.index, normalized_v1, marker='o', linestyle='-', color='b', label='Normalized Series')
plt.title('Normalized Series 1')
plt.xlabel('Index')
plt.ylabel('Normalized Value')
plt.legend()
plt.grid(True)
plt.show()
```
